## Spacetime

We declare the spacetime manifold $M$:

```python
M = Manifold(4, 'M')
print(M)
```

The Schwarzschild-Droste coordinates $(t,r,\theta,\phi)$:

```python
X.<t,r,th,ph> = M.chart(r't r:(0,+oo) th:(0,pi):\theta ph:(0,2*pi):\phi')
X
```

The Schwarzschild metric:

```python
g = M.lorentzian_metric('g')
m = var('m') ; assume(m>=0)
g[0,0], g[1,1] = -(1-2*m/r), 1/(1-2*m/r)
g[2,2], g[3,3] = r^2, (r*sin(th))^2 
g.display()
```

## Radial null geodesics

The outgoing family:

```python
var('v')
outgeod = M.curve({X: [r + 2*m*ln(abs(r/(2*m)-1)) + v, r, pi/2, pi]}, (r, 0, +Infinity))
outgeod.display()
```

The ingoing family:

```python
var('u')
ingeod = M.curve({X: [-r - 2*m*ln(abs(r/(2*m)-1)) + u, r, pi/2, pi]}, (r, 0, +Infinity))
ingeod.display()
```

```python
graph = Graphics()
for v0 in range(-20, 20, 2):
    graph += outgeod.plot(ambient_coords=(r,t), prange=(0.01, 1.99), parameters={m: 1, v: v0}, 
                          color='green', style='-', thickness=1, label_axes=False)
    graph += outgeod.plot(ambient_coords=(r,t), prange=(2.01, 8), parameters={m: 1, v: v0}, 
                          color='green', style='-', thickness=1, label_axes=False)
    graph += ingeod.plot(ambient_coords=(r,t), prange=(0.01, 1.99), parameters={m: 1, u: v0}, 
                          color='blue', style='--', thickness=1, label_axes=False)
    graph += ingeod.plot(ambient_coords=(r,t), prange=(2.01, 8), parameters={m: 1, u: v0}, 
                          color='blue', style='--', thickness=1, label_axes=False)
show(graph, axes_labels=[r"$r$", r"$t$"], aspect_ratio=1, ymin=-4, ymax=4)
```

Plot some light cones

```python
fout(r,v) = r + 2*ln(abs(r/2-1) + 1e-10) + v
fin(r,u) = -r - 2*ln(abs(r/2-1) + 1e-10) + u

def cone1(r,t):
    return t<fout(r,2) and t>fin(r,0) and r>1
def cone2(r,t):
    return t>fout(r,2) and t>fin(r,0) and t<1.8
def cone3(r,t):
    return t>fout(r,-2) and t>fin(r,4) and t<1.8
def cone4(r,t):
    return t>fout(r,-6) and t>fin(r,8) and t<1.8

graph += region_plot(cone1, (1,2), (0,2), incol='orange')
graph += region_plot(cone2, (2,8), (0,2), incol='orange')
graph += region_plot(cone3, (2,8), (0,2), incol='orange')
graph += region_plot(cone4, (2,8), (0,2), incol='orange')

show(graph, axes_labels=[r"$r$", r"$t$"], aspect_ratio=1, ymin=-4, ymax=4)
```

```python
graph.save("sch_rad_null_geod.png", aspect_ratio=1, ymin=-4, ymax=4)
```

## Eddington-Finkelstein coordinates

The ingoing Eddington-Finkelstein chart:

```python
X_EF.<te,r,th,ph> = M.chart(r'te:\tilde{t} r:(0,+oo) th:(0,pi):\theta ph:(0,2*pi):\phi') 
X_EF
```

```python
X_to_X_EF = X.transition_map(X_EF, [t+2*m*ln(abs(r/(2*m)-1)), r, th, ph])
X_to_X_EF.display()
```

```python
X_to_X_EF.inverse().display()
```

```python
ingeod.display()
```

```python
outgeod.display()
```

Plot of the radial null geodesics in terms of the ingoing Eddington-Finkelstein coordinates:

```python
graph = Graphics()
for v0 in range(-20, 20, 2):
    graph += outgeod.plot(chart=X_EF, ambient_coords=(r,te), prange=(0.01, 1.99), 
                          parameters={m: 1, v: v0}, color='green', style='-', 
                          thickness=1, label_axes=False)
    graph += outgeod.plot(chart=X_EF, ambient_coords=(r,te), prange=(2.01, 8), 
                          parameters={m: 1, v: v0}, color='green', style='-', 
                          thickness=1, label_axes=False)
    graph += ingeod.plot(chart=X_EF, ambient_coords=(r,te), prange=(0.01, 8), 
                         parameters={m: 1, u: v0}, color='blue', style='--', 
                         thickness=1, label_axes=False)
show(graph, axes_labels=[r"$r/m$", r"$\tilde{t}/m$"], aspect_ratio=1, ymin=-4, ymax=4)
```

```python
fout(r,v) = r + 4*ln(abs(r/2-1) + 1e-10) + v
fin(r,u) = -r + u

def cone1(r,t):
    return t<fout(r,4) and t>fin(r,2) and t<2
def cone2(r,t):
    return t>fout(r,2) and t>fin(r,0) and t<1.8
def cone3(r,t):
    return t>fout(r,-2) and t>fin(r,4) and t<1.5
def cone4(r,t):
    return t>fout(r,-10) and t>fin(r,8) and t<1.8

graph += region_plot(cone1, (1,2), (0,2), incol='orange')
#graph += region_plot(cone2, (2,8), (0,2), incol='yellow')
graph += region_plot(cone3, (2,8), (0,2), incol='orange')
graph += region_plot(cone4, (2,8), (0,2), incol='orange')

show(graph, axes_labels=[r"$r$", r"$\tilde{t}$"], aspect_ratio=1, ymin=-4, ymax=4)
```

```python
graph.save("sch_rad_null_geod_EF.png", aspect_ratio=1, ymin=-4, ymax=4)
```

The outgoing Eddington-Finkelstein chart:

```python
X_EF.<tr,r,th,ph> = M.chart(r'tr:t* r:(0,+oo) th:(0,pi):\theta ph:(0,2*pi):\phi') 
X_EF
```

```python
X_to_X_EF = X.transition_map(X_EF, [t-2*m*ln(abs(r/(2*m)-1)), r, th, ph])
X_to_X_EF.display()
```

```python
X_to_X_EF.inverse().display()
```

```python
ingeod.display()
```

```python
outgeod.display()
```

Plot of the radial null geodesics in terms of the outgoing Eddington-Finkelstein coordinates:

```python
graph = Graphics()
for v0 in range(-20, 20, 2):
    graph += outgeod.plot(chart=X_EF, ambient_coords=(r,tr), prange=(0.01, 1.99), 
                          parameters={m: 1, v: v0}, color='green', style='-', 
                          thickness=1, label_axes=False)
    graph += outgeod.plot(chart=X_EF, ambient_coords=(r,tr), prange=(2.01, 8), 
                          parameters={m: 1, v: v0}, color='green', style='-', 
                          thickness=1, label_axes=False)
    graph += ingeod.plot(chart=X_EF, ambient_coords=(r,tr), prange=(0.01, 8), 
                         parameters={m: 1, u: v0}, color='blue', style='--', 
                         thickness=1, label_axes=False)
show(graph, axes_labels=[r"$r$", r"$t^*$"], aspect_ratio=1, ymin=-4, ymax=4)
```

```python
fout(r,v) = r + v
fin(r,u) = -r - 4*ln(abs(r/2-1) + 1e-10) + u

def cone1(r,t):
    return t<fin(r,2) and t>fout(r,2) and t<3.5
def cone2(r,t):
    return t>fout(r,2) and t>fout(r,0) and t<1.8
def cone3(r,t):
    return t>fin(r,4) and t>fout(r,-2) and t<3
def cone4(r,t):
    return t>fin(r,4) and t>fout(r,-8) and t<-2

graph += region_plot(cone1, (0,2), (0,4), incol='orange')
#graph += region_plot(cone2, (2,8), (0,2), incol='yellow')
graph += region_plot(cone3, (3,5), (0,3), incol='orange')
graph += region_plot(cone4, (4,8), (-3,-2), incol='orange')


show(graph, axes_labels=[r"$r$", r"$t^*$"], aspect_ratio=1, ymin=-4, ymax=4)
```
